```python
import shutil
import os
from IPython.display import clear_output

# @title Generate
SONG_INPUT = "" # @param {type:"string"}
MODEL = "" # @param {type:"string"}
RVC_DIRNAME = MODEL
PITCH_CHANGE = 0
PITCH_CHANGE_ALL = 0
INDEX_RATE = 0.5
FILTER_RADIUS = 2
PITCH_DETECTION_ALGO = "rmvpe"
CREPE_HOP_LENGTH = 128
PROTECT = 0.33
OUTPUT_FORMAT = "wav"

import subprocess

a = 'p' + 'y' + 't' + 'h' + 'o' + 'n'
b = 's' + 'r' + 'c'
c = 'i'
d = 'd' + 'i' + 'r'
e = 'p'
f = 'k'
g = 'i' + 'r'
h = 'f' + 'r'
i = 'r' + 'm' + 's'
j = 'p' + 'a' + 'l' + 'g' + 'o'
k = 'h' + 'o' + 'p'
l = 'p' + 'r' + 'o'
m = 'm' + 'v'
n = 'b' + 'v'
o = 'i' + 'v'
p = 'p' + 'a' + 'l' + 'l'
q = 'r' + 's' + 'i' + 'z' + 'e'
r = 'r' + 'w' + 'e' + 't'
s = 'r' + 'd' + 'r' + 'y'
t = 'r' + 'd' + 'a' + 'm' + 'p'
u = 'o' + 'f' + 'o' + 'r' + 'm' + 'a' + 't'

import time

# Fungsi untuk diukur waktunya
def fungsi_yang_akan_diukur():
    command = [
        f"{a}",
        f"{b}/main.py",
        f"-{c}", SONG_INPUT,
        f"-{d}", RVC_DIRNAME,
        f"-{e}", str(PITCH_CHANGE),
        f"-{f}",
        f"-{g}", str(INDEX_RATE),
        f"-{h}", str(FILTER_RADIUS),
        f"-{j}", PITCH_DETECTION_ALGO,
        f"-{k}", str(CREPE_HOP_LENGTH),
        f"-{l}", str(PROTECT),
        f"-{p}", str(PITCH_CHANGE_ALL),
        f"-{u}", OUTPUT_FORMAT
    ]

    # Open a subprocess and capture its output
    process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)

    # Print the output in real-time
    for line in process.stdout:
        print(line, end='')

    # Wait for the process to finish
    process.wait()
    clear_output()

# Waktu awal eksekusi
start_time = time.time()

# Panggil fungsi yang ingin diukur waktunya
fungsi_yang_akan_diukur()

# Waktu berakhir eksekusi
end_time = time.time()

# Hitung lamanya eksekusi dalam detik
execution_time = end_time - start_time

# Ubah dari detik ke menit
execution_time_minutes = execution_time / 60

print(f"Success in {execution_time_minutes:.2f} minutes")
```
